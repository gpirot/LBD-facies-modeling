### Overview
This Jupyter notebook works with driller logs from the state of Queensland. The code reads in a file with drill logs for 1000s of bores in Queensland and uses rules generated by the user to produce four classifications:

1. P1 the primary material identified (i.e., sand, clay, basement)
2. P2 any other primary materials (i.e. sand and clay would produce clay as P2)
3. Material descriptors (i.e., sandy, weathered) - Could relate to P1 or P2
4. Grainsize (i.e., fine, medium, coarse) - Could relate to P1 or P2
5. Colours. The code will identify the first two colours (if present)

WITH ANY DESCRIPTOR, MATERIAL, ETC., <b>ALWAYS USE ALL CAPS,</b> 

### The input file
The input file was generated from the state-wide file from the Australian Groundawter Explorer and the shapefile of the Lower Burdekin Water area in QGIS. The code could be used to classify every bore in Queensland (using the original STRATA_LOG.txt file), and/or work with other regions with some minor adjustments to the code.

The file includes the original data from the driller log file for Queensland, with the following columns added: latitude, longitude, easting, northing and the easting/northing zone (i.e., 55). This is a slightly over the top approach, and could have been included later, but it's not an issue.

It should be checked if the input file contains all of the bores that are in the Queensland database. If not, a new input file is required.

### Output files
Two files are produced from the codes that are run here. In both instances, the the user selects this file name in the relevant cell. In the preliminary example, these file names are LBW_STRATA_LOG_processed.csv and LBW_STRATA_LOG_summary.csv

1. A file that attempts to classify each layer from drill logs. For each layer, the code classifies the driller descriptions according to the 5 outputs listed in the overview.
2. A summary of the wells, reporting the overall thickness as well as a 'confidence' value. The confidence is calculated from the total number of characters for a bore, and the overall bore length. The score is presented as 

score = (n characters/ total bore length)

which is then normalised by the max score, thus the scores range from 1 to ~0.

#### Step 7: Clean up

The drill logs often contain descriptors that we do not require.  When these are identified, the code typically produces "OTHER NOT CLASSIFIED". We can search through these and the original descriptions to find layers to remove from the result above.

Here we run another function called tidy_df

- <b> P1_value </b> This is the P1 search term that the code looks through
- <b> descr_phrases </b> A list of terms to look for. If encountered, the line will be deleted.

```python
print('Working...')
# terms used to search df2 and delete any unwanted outputs
P1_value = 'OTHER NOT CLASSIFIED'
descr_phrases = ['BOTTOM OF HOLE', 'EOH', 'BOH', 'AIRLIFTED']

# and just some temporary rules to classify some of the "OTHER NOT CLASSIFIED" values
# Condition 1: If P1 is "OTHER NOT CLASSIFIED" and "WEATHERED" is in Material_Descriptor, set P1 to "Basement"
condition_1 = (df2['P1'] == 'OTHER NOT CLASSIFIED') & (df2['Material_Descriptor'].str.contains('WEATHERED'))
df2.loc[condition_1, 'P1'] = 'BASEMENT'

# Condition 2a: If P1 is "OTHER NOT CLASSIFIED" and grain_size includes "COARSE", "MEDIUM", etc., set P1 to "SAND"
grain_size_keywords = ['COARSE', 'MEDIUM', 'MED', 'FINE'] # Add other similar keywords as needed
condition_2a = (df2['P1'] == 'OTHER NOT CLASSIFIED') & (df2['grain_size'].str.contains('|'.join(grain_size_keywords)))
df2.loc[condition_2a, 'P1'] = 'SAND'

# Condition 2b: If P1 is "OTHER NOT CLASSIFIED" and Material_Descriptor includes "SANDY", set P1 to "SAND"
condition_2b = (df2['P1'] == 'OTHER NOT CLASSIFIED') & (df2['Material_Descriptor'].str.contains('SANDY'))
df2.loc[condition_2b, 'P1'] = 'SAND'

# Condition 3: If P1 is "OTHER NOT CLASSIFIED" and "FIRM" or "CLAYEY" is in DESCR, set P1 to "CLAY"
clay_keywords = ['FIRM', 'HARD', 'CLAYEY'] 
condition_1 = (df2['P1'] == 'OTHER NOT CLASSIFIED') & (df2['DESCR'].str.contains('|'.join(clay_keywords)))
df2.loc[condition_1, 'P1'] = 'CLAY'

# run the tidy_df function
df3 = tidy_df(df2, P1_value, descr_phrases, output_filename)

print(str(output_filename)+ ' saved')
df3.head()
```
